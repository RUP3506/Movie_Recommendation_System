```python
movie  = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
credit = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
#lets merge the dat based on title
data  = movie.merge(credit, on='title')
data.head(2)
```

```python
df = data[['movie_id','title','overview','genres','keywords','cast','crew']]
df.head()
```

```python
for i in df.columns:
    print(i,":")
    print(df[i][0])
    print("\n")
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

Create tags :

```python
df['overview'] = df['overview'].apply(lambda x : x.split())

# lets create a function to extract the details of genres and keyword
def name_extract(obj):
    l = []
    for i in eval(obj):
        l.append(i['name'])
    return l

df['genres'] = df['genres'].apply(lambda x: name_extract(x)) 
df['keywords'] = df['keywords'].apply(name_extract)

# lets create a function to extract the top 5 names from cast
def extract_5_names(obj):
    l = []
    n = 0
    for i in eval(obj):
        if n<5:
            l.append(i['name'])
            n += 1
        else :
            break
    return l

df['cast'] = df['cast'].apply(extract_5_names)
df.cast[0]
```

```python
def extract_director_name(obj):
    l = []
    for i in eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
    return l

df['crew'] = df['crew'].apply(extract_director_name)    
```

```python
df.head(2)
```

```python
# now lets remove the spaces between words like names ...

df['genres'] = df['genres'].apply(lambda x: [i.replace(' ','') for i in x])
df['keywords'] = df['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
df['cast'] = df['cast'].apply(lambda x: [i.replace(' ','') for i in x])
df['crew'] = df['crew'].apply(lambda x: [i.replace(' ','') for i in x])
df.head()
```

```python
df['tags'] = df['overview']+df['genres']+df['keywords']+2*df['cast']+ 5*df['crew']
new_df = df[['movie_id','title','tags']]
```

```python
df.tags
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : ' '.join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')

vector = cv.fit_transform(new_df['tags']).toarray()
vector
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity[0].shape
```

```python
new_df.head(3)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df['title'][i[0]])
recommend('Avatar')
```

Now lets save the files in a pickle file ...

```python
import pickle
pickle.dump(similarity,open('similarity.pkl','wb'))
pickle.dump(new_df.to_dict(),open('movies_dict.pkl','wb'))
```
